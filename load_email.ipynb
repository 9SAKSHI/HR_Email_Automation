```python
import pandas as pd
import win32com.client

def load_and_replace_template(template_path, candidate_name, date_of_joining, location):
    try:
        outlook = win32com.client.Dispatch("Outlook.Application")
        mail_item = outlook.CreateItemFromTemplate(template_path)
        body = mail_item.Body
        
        body = body.replace("{{candidate_name}}", candidate_name)
        body = body.replace("{{date_of_joining}}", date_of_joining)
        body = body.replace("{{location}}", location)
        
        # Print or return the modified body to view
        print("Modified Body:")
        print(body)
        
        # You can set the modified body back to the mail item if needed
        mail_item.Body = body
        
        # Send the email
        mail_item.Send()
        
        print(f"Email sent to {candidate_name}")
        
    except Exception as e:
        print(f"Error loading and replacing the template: {e}")

# def process_excel_and_send_emails(excel_file_path, template_path):
#     # Read the Excel file into a pandas DataFrame
#     df = pd.read_excel(excel_file_path)

#     # Loop through each row in the DataFrame
#     for index, row in df.iterrows():
#         # Extract the required information from each row
#         candidate_name = row['Name']  # Assuming 'Name' is the column for candidate name
#         date_of_joining = row['DOJ']  # Assuming 'DOJ' is the column for date of joining
#         location = row['Location']  # Assuming 'Location' is the column for location
        
#         # Call the function to load the template and replace the placeholders
#         load_and_replace_template(template_path, candidate_name, date_of_joining, location)

# # Example usage
# excel_file_path = r'C:\Users\SNR23\Desktop\HR Automation\Email_Automation\Gearup Email Automation\Gearup Email Automation\DSGS_Offer status sheet_2025_V1.xlsx'
# template_path = r"C:\Users\SNR23\Desktop\_test.msg"

# # Process the Excel file and send emails
# process_excel_and_send_emails(excel_file_path, template_path)
```

```python
def process_excel_and_send_emails(excel_file_path, template_path):
    # Read the Excel file into a pandas DataFrame
    df = pd.read_excel(excel_file_path)

    # Loop through each row in the DataFrame
    for index, row in df.iterrows():
        # Extract the required information from each row
        candidate_name = row['Name']  # Assuming 'Name' is the column for candidate name
        date_of_joining = row['DOJ']  # Assuming 'DOJ' is the column for date of joining
        location = row['Location']  # Assuming 'Location' is the column for location
        
        # Call the function to load the template and replace the placeholders
        load_and_replace_template(template_path, candidate_name, date_of_joining, location)

# Example usage
excel_file_path = r'C:\Users\SNR23\Desktop\HR Automation\Email_Automation\Gearup Email Automation\Gearup Email Automation\DSGS_Offer status sheet_2025_V1.xlsx'
template_path = r"C:\Users\SNR23\Desktop\_test.msg"

# Process the Excel file and send emails
process_excel_and_send_emails(excel_file_path, template_path)
```
